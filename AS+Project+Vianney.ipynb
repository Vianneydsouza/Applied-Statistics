###Q2: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

```python
#2.A. Probability that none of the items are defective?
#2.B. Probability that exactly one of the items is defective?

#importing binom so that we can use binom.pmf function
from scipy.stats import binom
p = 0.05
n = 10
k = 0

for k in range (0,2):
  prob = binom.pmf(k=k, n=n, p=p)

  print(f"Probability that", k ,"of the items are defective:", round(prob, 4))
```

*A. The Probability that none of the items are defective is 0.5987 or 59.87%*

*B. The Probability that exactly one of the items is defective is 0.3151 or 31.51%*

```python
#2.C. Probability that two or fewer of the items are defective?
p = 0.05
n = 10
k = 0
prob_two_or_fewer = 0

for k in range (0,3):
  prob = binom.pmf(k=k, n=n, p=p)
  prob_two_or_fewer += prob  #adding the probabilities

print(f"Probability that two or fewer of the items are defective:", round(prob_two_or_fewer, 4))
```

*C. The Probability that two or fewer of the items are defective is 0.9885 or 98.85%*

```python
#2.D. Probability that three or more of the items are defective ?
p = 0.05
n = 10
k = 0
prob_three_or_more = 0

for k in range (0,3):
  prob = binom.pmf(k=k,n=n,p=p)
  prob_three_or_more += prob
  probability = 1 - prob_three_or_more

print(f'Probability that three or more of the items are defective:', round(probability, 4))
```

*D. The Probability that three or more of the items are defective is 0.0115 or 1.15%*

###Q3: A car salesman sells on an average 3 cars per week.

```python
#3.A. What is Probability that in a given week he will sell some cars?
from scipy.stats import poisson
rate = 3 # average sells 3 cars per week
n = np.arange(0,8)
poisson = 1 - stats.poisson.pmf(n, rate)
print(f'The Probability that in a given week he will sell some cars:', round(poisson[0], 4))
```

*A: The Probability that in a given week he will sell some cars is 0.9502 or 95.02%*

```python
#3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?
rate = 3
prob = 0
n = np.arange(2,5)
poisson = stats.poisson.pmf(n, rate)
prob = sum(poisson)
print(f'The Probability that in a given week he will sell 2 or more but less than 5 cars will be:', round(prob, 4))
```

*B: The Probability that in a given week he will sell 2 or more but less than 5 cars will be: 0.616 or 61.61%*


```python
#3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.
rate = 3
n = np.arange(0,16) # considering 15 cars sold per week
poisson = stats.poisson.pmf(n,rate)
prob = np.cumsum(poisson)
prob
plt.figure(figsize=(10,6))
sns.lineplot(x=n, y=prob, marker='o', color='red')
plt.xlabel('Number of Cars sold per week')
plt.ylabel('Probability of cars sold per week')
plt.title('Poisson');
```

###Q4: Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours independently place orders with the same bot. Answer the following questions.

```python
#4.A. What is the probability that all three orders will be recognised correctly?
#4.B. What is the probability that none of the three orders will be recognised correctly?

n = 3
p = 0.868

for k in range(0,4):
  prob = binom.pmf(k=k,n=n,p=p)
  print(f'The probability that {k} of the three orders will be recognised correctly is:', round(prob, 4))
```

*A: The probability that all three orders will be recognised correctly is 0.654 or 65.4%*

*B: The probability that none of the three orders will be recognised correctly is 0.0023 or 0.23%*

```python
#4.C. What is the probability that at least two of the three orders will be recognised correctly?
n = 3
p = 0.868
prob_atleast_two = 0
for k in range (2, n + 1): # we need atleast 2 sucessess
  prob = binom.pmf(k=k,n=n,p=p)
  prob_atleast_two += prob
print(f'The probability that at least two of the three orders will be recognised correctly:', round(prob_atleast_two, 4))
```

*C: The probability that at least two of the three orders will be recognised correctly is 0.9523 or 95.23%*

A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

```python
#importing libraries
from scipy.stats import stats
from scipy.stats import ttest_ind
```

```python
data.head()
```

```python
# dividing into groups
operating = data['Funds_in_million'][data['OperatingState']=='Operating']
closed = data['Funds_in_million'][data['OperatingState']=='Closed']

tstats, p_value = ttest_ind(operating, closed, equal_var=True)  #performing ttest
print(tstats, p_value)
```

```python
#taking alpha as 0.05
alpha = 0.5
if p_value < alpha:
  print('there is significant difference between Funds raised by companies that are still operating vs companies that closed down')
else:
  print('there is no significant difference between Funds raised by companies that are still operating vs companies that closed down')
```

B. Write the null hypothesis and alternative hypothesis.

```python
Null hypothesis (H0) will be there is no significant difference between Funds raised by companies that are still operating and companies that closed down.
Alternative hypothesis (H1) will be there is significant difference between Funds raised by companies that are still operating and companies that closed down.
```
